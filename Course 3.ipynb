## Project B

```python
from sympy import symbols, solve, N

r = symbols('r')
solutions = solve(100/(1+r) + 200/((1+r)**2) + 300/((1+r)**3) + 400/((1+r)**4) + 500/((1+r)**5) - 1000, r)

# To get the numerical approximations of the roots
numeric_solutions = [sol.evalf() for sol in solutions]
print(numeric_solutions)
```
